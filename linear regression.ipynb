```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
np.random.seed(0)
E_S = np.random.normal(4.0,0.5,100)
Price = 1000+E_S*1000+np.random.normal(0,2000,100)
```

```python
E_S
```

```python
Price
```

```python
df=pd.DataFrame({"Engine Size":E_S,"Price":Price})
```

```python
df
```

```python
X=df[["Engine Size"]]
```

```python
X
```

```python
Y=df[["Price"]]
```

```python
Y
```

```python
model = LR()
model.fit(X,Y)
```

```python
y=model.predict(X)
y
```

```python
MSE=mean_squared_error(Y,y)
MAE=mean_absolute_error(Y,y)
R2=r2_score(Y,y)
```

```python
MSE
```

```python
MAE
```

```python
R2
```

```python
plt.figure(figsize=(10, 5))

plt.scatter(X, Y, color='blue', label="Actual Data")
plt.plot(X, y, color='red', label="Predicted Data")
plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
np.random.seed(0)
E_S=np.random.normal(3.0,0.5,100)
w_b=np.random.normal(2.0,0.5,100)
amt=1000+E_S*1000+np.random.normal(0,2000,100)
```

```python
df2=pd.DataFrame({"Engine size":E_S,"Wheel Base": w_b, "Amonunt": amt})
```

```python
df2
```

```python
X1=df2[["Engine size", "Wheel Base"]]
```

```python
Y1=df2["Amonunt"]
```

```python
MLRModel=LR()
```

```python
MLRModel.fit(X1,Y1)
```

```python
y1=MLRModel.predict(X1)
```

```python
y1
```

```python
MSE1=mean_squared_error(Y1,y1)
MAE1=mean_absolute_error(Y1,y1)
R21=r2_score(Y1,y1)
```

```python
x_surf, y_surf = np.meshgrid(
    np.linspace(
        df2['Engine size'].min(),
        df2['Engine size'].max(),
        100
    ),
    np.linspace(
        df2['Wheel Base'].min(),
        df2['Wheel Base'].max(),
        100
    )
)
```

```python
x_surf
```

```python
y_surf
```

```python
xy_surf=np.c_[x_surf.ravel(),y_surf.ravel()]
```

```python
xy_surf
```

```python
z_surf=MLRModel.predict(xy_surf).reshape(x_surf.shape)
```

```python
z_surf
```

```python
fig = plt.figure()
ax=fig.add_subplot(111,projection ='3d')
ax.scatter(df2["Engine size"],df2["Wheel Base"], df2["Amonunt"],color='blue' ,label="Actual data")
ax.plot_surface(x_surf,y_surf,z_surf,color='red',label="Predicted Value", alpha=0.5)
ax.set_xlabel("Engine size")
ax.set_ylabel("Wheel Base")
ax.set_zlabel("Price")
ax.set_title("Engine Size and Wheel Base vs Price")
plt.legend()
plt.show()
```
