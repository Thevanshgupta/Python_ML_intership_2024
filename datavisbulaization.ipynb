```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sm=pd.read_csv('student_marks.csv')
sm
```

```python
sm['Total']=sm[['English','Hindi','Maths','Science','SST']].sum(axis=1)
```

```python
sm
```

```python
sm['Percentage']=sm['Total']/5
```

```python
sm
```

A=100-80
B=60-80
C=60-40
D=40-30
F=0-30

```python
def assign_grade(percentage):
    if percentage >= 80 :
        return 'A'
    elif percentage >= 60:
        return 'B'
    elif percentage >= 40:
        return 'C'
    elif percentage >= 30:
        return 'D'
    else:
        return 'F'
sm['Grade'] = sm['Percentage'].apply(assign_grade)
print(sm)
```

```python
plt.figure(figsize=(10,6))

for sub in ['English', 'Hindi', 'Maths', 'Science', 'SST']:
    plt.plot(sm['Name'], sm[sub], marker='o', label=sub)
    plt.title("Marks in Different Subjects")
    plt.xlabel('Name')
    plt.ylabel('Marks')
    plt.legend()
    plt.grid(True)
    plt.show()
```
